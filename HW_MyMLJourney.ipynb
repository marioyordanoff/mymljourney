# Защо избрах да уча машинно обучение

##### Интереса ми към изкуствения интелект и машинното обучение се породи в мен преди около 3 години когато все още бях в гимназията. Машинното обучение е трудно умение за усъвършенстване, но за сметка на това има ти дава сили да имаш огромно влияние върху развитиетието на технологиите.
